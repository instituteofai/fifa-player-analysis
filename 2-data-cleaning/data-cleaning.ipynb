```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.read_csv('../data/raw_data.csv', index_col=0)
df.head()
```

## height
First, we convert the height to inches.

```python
df['height'] = list(map(lambda s: int(s.split('\'')[0])*12 + int(s.split('\'')[1].split('"')[0]), df['height']))
```

```python
df.head()
```

## weight

We remove the `lbs` suffix from weight.

```python
df['weight'] = list(map(lambda x: x.split('lbs')[0], df['weight']))
```

```python
df.head()
```

## value
We remove all special characters from value.

First, we investigate which extra characters exist so that we can decide how to treat them.

```python
unique_chars = set()
for value in df['value']:
    for c in value:
        unique_chars.add(c)
```

```python
unique_chars
```

We can deal with the special characters as follows:
 - Drop the euro symbol
 - If there is `K` then multiply the value by 1 000
 - If there is 'M' then multiply the value by 1 000 000
